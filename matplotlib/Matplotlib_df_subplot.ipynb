```python
import matplotlib.pyplot as plt
from pylab import  text
import pandas as pd
plt.rcParams["figure.figsize"] = 12, 4
users_df=pd.read_csv("users.dat",delimiter="::",engine='python',names =['UserID','Gender','Age','Occupation','Zip-code'],header=None)
```

```python
def view_columns_desc(df):
    for columns in   df.select_dtypes(include=['int64','float64']).columns:
        fig = plt.figure()
        ax1 = fig.add_subplot(131)
        ax2 = fig.add_subplot(132)
        ax3 = fig.add_subplot(133)
        df[[columns]].boxplot(ax=ax1)
        df[[columns]].plot(kind='hist',bins=20,ax=ax2)
        data=df[[columns]].describe()
        data=data.append(pd.DataFrame([len(df[columns].unique())], columns = [columns], index=['Unique']) )
        text(0.5,0.5,str(data), horizontalalignment='center',verticalalignment='center',transform = ax3.transAxes)
        plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
df_new=users_df[['Age','Occupation']]
view_columns_desc(df_new)
```
